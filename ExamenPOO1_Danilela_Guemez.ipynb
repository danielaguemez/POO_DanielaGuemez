```python
data8 = pd.DataFrame(np.arange(12).reshape((3, 4)),index=['CDMX', 'Guanajuato', 'Puebla'],columns=['uno', 'dos', 'tres', 'cuatro'])

data8
```

El data frame tiene diversas formulas:
La fórmula de `np.arange(12)` significa que va a darte un arreglo de 12 caracteres desde el 0 al 11, la fórmula `.reshape((3,4))` significa que te va a dar este arreglo en una matriz de 3x4, la fórmula de `index=['CDMX', 'Guanajuato', 'Puebla']` se refiere a que esos van a ser los títulos de las filas, mientras que `columns=['uno', 'dos', 'tres', 'cuatro']` se refiere a que esos van a ser los títulos de las columnas

Primero se debe de poner el nombre la columna en comillas `'key1'` y posteriormente dos puntos `:` y adentro de corchetes `[]` los datos, ya sean numéricos o no. Por ejemplo, dentro de `key1` y `key2` hay caracteres, mientras que en `data1` y `data2` hay numeros aleatorios que se generan con `np.random.randn(5)` que esto significa que generara cinco números aleatorios que siguen una distribución normal.
